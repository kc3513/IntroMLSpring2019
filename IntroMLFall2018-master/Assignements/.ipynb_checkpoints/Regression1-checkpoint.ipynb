```python
import pandas as pd
import numpy as np
pd.core.common.is_list_like = pd.api.types.is_list_like

from pandas_datareader import data as pdr

import matplotlib.pyplot as plt

import fix_yahoo_finance as yf
yf.pdr_override() # <== that's all it takes :-)

# download dataframe
data1 = pdr.get_data_yahoo("SPY", start="2017-01-01", end="2017-04-30")

# download Panel
data = pdr.get_data_yahoo(["SPY", "IWM"], start="2017-01-01", end="2017-04-30")



close = data1['Close']

all_weekdays = pd.date_range(start="2017-01-01", end="2017-04-30", freq='B')

# How do we align the existing prices in adj_close with our new set of dates?
# All we need to do is reindex close using all_weekdays as the new index
close = close.reindex(all_weekdays)

close = close.fillna(method='ffill')


msft = close#.loc['SPY']

# Calculate the 20 and 100 days moving averages of the closing prices
short_rolling_msft = msft.rolling(window=20).mean()
long_rolling_msft = msft.rolling(window=100).mean()

# Plot everything by leveraging the very powerful matplotlib package
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(msft.index, msft, label='SPY')
ax.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling')
ax.plot(long_rolling_msft.index, long_rolling_msft, label='100 days rolling')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()

#plt.show()


msft = msft.dropna()
#msft_norm = (msft - msft.mean()) / (msft.max() - msft.min())


tmp = msft.index
numDays = tmp.shape[0]

days = np.arange(1,numDays+1)

# normalized prices
print(msft.index)


tmp1 = 80


X_train = days[0:tmp1]
X_test = days[tmp1:]

y_train = msft[0:tmp1]
y_test = msft[tmp1:]

print(y_train)
print(X_train)
# y_train = msft[days < tmp1]
# y_test = msft[days >= tmp1]


# print(days)
# print(days < tmp1)
# #print(y_train)
# print(X_test)
#print(y_test)



from sklearn.linear_model import LinearRegression

print(X_train.transpose)
print(y_train.values)

print(X_train.shape)
tmp = y_train.values
print(tmp.shape)

lr = LinearRegression().fit(X_train.reshape(-1,1),tmp.reshape(-1,1))

pred = lr.predict(X_test.reshape(-1,1))


# #training a linear prediction model

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(days, msft, label='SPY')
ax.plot(X_train.reshape(-1,1), tmp.reshape(-1,1), label='training')
ax.plot(X_test, pred, label='test')

# ax.set_xlabel('Date')
# ax.set_ylabel('Adjusted closing price ($)')
# ax.legend()

plt.show()



```
